# Step 1: Load the given Banknote authentication dataset.

```python
import pandas as pd
df = pd.read_csv('Banknote_authentication_dataset.csv')
```

# Step 2: Calculate statistical measures, e.g. mean and standard deviation.

```python
print("Mean for V1 is", np.mean(df["V1"]))
print("Mean for V2 is", np.mean(df["V2"]))
print("Standard Deviation for V1 is", np.std(df["V1"]))
print("Standard Deviation for V2 is", np.std(df["V2"]))
```

# Step 3: Visualise your data as you consider fit.

```python
import matplotlib.pyplot as plt
V1 = df["V1"]
V2 = df["V2"]

plt.figure(figsize = (12,10))
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(V1,V2)
```

# Step 4: Evaluate if the given dataset is suitable for the K-Means clustering task.

```python
from sklearn.cluster import KMeans
V12 = np.column_stack((V1,V2))
res = KMeans(n_clusters= 3).fit(V12)
clusters = res.cluster_centers_
plt.figure(figsize = (12,10))
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(V1,V2)
plt.scatter(clusters[:,0],clusters[:,1],s = 1000,alpha=0.6)

plt.grid()
plt.show()
```
